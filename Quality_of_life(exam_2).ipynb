# **7. Basic Line/Bar Graph**

```python
# Graph the Quality of life of United the sates
fig = px.line(usa, x= 'Year_str',y ='Quality of Life Index')

# Change title
fig.update_layout(title = 'Quality of live in USA',
                  title_font_size = 30,
                  title_x = 0.5)

# Change Axis
fig.update_xaxes(title = 'Dates', title_font_size = 20)
fig.update_yaxes(title = 'Life in years', title_font_size = 20)

fig.show()
```

```python
# Graph the Quality of life of United the sates
fig = px.bar(usa, x= 'Year_str',y ='Quality of Life Index')

# Change title
fig.update_layout(title = 'Quality of Life in USA',
                  title_font_size = 30,
                  title_x = 0.5,
                  yaxis_range = [70,80])

# Change Axis
fig.update_xaxes(title = 'Dates', title_font_size = 20)
fig.update_yaxes(title = 'Life in years', title_font_size = 20)
fig.show()

```

## Two y-axis 

```python
# Create a figure with secondary axis
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x = usa['Year_str'], y = usa['Quality of Life Index'], name = 'Purchasing Power Index'),
    secondary_y = False, 
)

fig.add_trace(
    go.Scatter(x = usa['Year_str'], y = usa['Safety Index'], name = 'Health Care Index'),
    secondary_y = True, 
)

fig.show()
```

```python
# Create a figure with secondary axis
fig = make_subplots(specs = [[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Bar(x = usa['Year_str'], y = usa['Quality of Life Index'], name = 'Purchasing Power Index'),
    secondary_y = False, 
)

fig.add_trace(
    go.Scatter(x = usa['Year_str'], y = usa['Safety Index'], name = 'Health Care Index'),
    secondary_y = True, 
)

fig.show()
```
